## Model 

Define the features for input and output and storing it into x and y

assiging the value of w,b = 0

```python
x = data[['age', 'bmi', 'children','discount_eligibility', 'gender', 'region']].to_numpy()
y = data['premium'].to_numpy()
```

```python
n_feature = x.shape[1]
w = np.zeros(n_feature)
b = 0 
```

```python
print(x.shape)
print(w.shape)
```

Declaring the cost function 

```python
mu = np.mean(x, axis=0)
sigma = np.std(x, axis=0)
x_scaled = (x - mu) / sigma
```

```python
def cost_function(x_scaled,y,w,b):
    x_scaled = (x - mu) / sigma
    m = x.shape[0]
    f_pred = np.dot(x_scaled,w) + b
    cost = np.sum((f_pred - y)**2)
    return ((cost)/(2*m))
```

```python
print(cost_function(x_scaled,y,w,b)) # intial value of cost function 
```

Now our Aim is the min(cost_function) - using partial derivatives 

```python
m = x.shape[0]
f_pred = np.dot(x_scaled,w) + b
error = f_pred - y
dj_dw = np.dot(x.T, error)/m
dj_db = np.sum(error)/m
```

```python
alpha = 0.1
epochs = 10000

for i in range(epochs):

    f_pred = np.dot(x_scaled, w) + b

    error = f_pred - y

    dj_dw = np.dot(x_scaled.T, error) / m
    dj_db = np.sum(error) / m

    w = w - alpha * dj_dw
    b = b - alpha * dj_db

    if i % 1000 == 0:
        print(i, cost_function(x_scaled, y, w, b))
```

```python
print("W : ",w)
print("B : ",b)
```

```python
y_pred = np.dot(x_scaled, w) + b
```

```python
rmse = np.sqrt(np.mean((y_pred - y)**2))
print(rmse)
```

```python
ss_res = np.sum((y- y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)

r2 = 1 - (ss_res / ss_tot)
print(r2)
```
